```python
import pandas as pd
d=pd.read_csv("/content/drive/MyDrive/diabetes[1].csv")
d.head()
```

```python
import seaborn as sns
sns.pairplot(d)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
d['location']=le.fit_transform(d['location'])
d['gender']=le.fit_transform(d['gender'])
d['frame']=le.fit_transform(d['frame'])
d['dp.2s']=le.fit_transform(d['bp.2s'])
d['dp.2d']=le.fit_transform(d['bp.2d'])
```

```python
corr_data=d.drop(columns=['weight'])
corr_data=corr_data.corr()
sns.heatmap(corr_data,annot=True,cmap='coolwarm',fmt='.2f',annot_kws={'size':.24})
```
